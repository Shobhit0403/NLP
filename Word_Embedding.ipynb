```python
from keras.preprocessing.text import one_hot
from keras.layers import Embedding
from keras.models import Sequential
from keras.preprocessing.sequence import pad_sequences
```

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
 ]
```

```python
sent 
```

```python
voc_size = 10000
```

```python
onehot_representation =[one_hot(word, voc_size) for word in sent]
print(onehot_representation)
```

```python
import numpy as np
```

```python
sent_lenght =8
embedded_docs =pad_sequences(onehot_representation, padding='pre', maxlen=sent_lenght)
print(embedded_docs)
```

```python
model = Sequential()

model.add(Embedding(voc_size, 10, input_length=sent_lenght))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
